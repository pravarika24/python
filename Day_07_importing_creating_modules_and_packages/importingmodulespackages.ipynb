```python
import math
#Now we can access all functions inside the `math` module.
print(math.sqrt(16))
import numpy
arr = numpy.array([1,2,3,4,5])
print(arr)
```

```python
#Importing Specific Functions
from math import sqrt, pi
print(sqrt(16))
print(pi)
#using alias
import numpy as np
numbers = np.array([1,2,3,4,5])
print(numbers)
#Importing All Functions
from math import *
print(sqrt(16))
print(pi)
```

```python
#Creating Custom Packages - 
# we have to create a folder(package) and we have to write __init__.py 
# I have created a folder named "package" and inside it a module maths.py that contains an addition function
from package.maths import addition
print(addition(2,3))
#Import the Entire Module
from package import maths
print(maths.addition(2,3))
#Import Everything
from package.maths import *
print(addition(2,3))
#Import from a Subpackage
from package.subpackages.mult import multiply
print(multiply(4,5))
```
