- Attributes (Variables)
- Methods (Functions)

```python
#Creating Your First Class
class Car:
    pass
audi = Car()
audi.windows = 4
print(audi.windows)
#Although valid, this is **not recommended**.
```

```python
#Best Practice:** Always initialize attributes using a constructor.
#Instance variables belong to individual objects.
#They are initialized inside the constructor.
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
dog1 = Dog("Buddy", 3)
print(dog1.name)
print(dog1.age)
#Another object.
dog2 = Dog("Lucy", 4)
print(dog2.name)
print(dog2.age)
```

```python
#Constructors (`__init__()`)
#The constructor (`__init__()`) is automatically called whenever an object is created.
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    #Instance Methods
    def bark(self):
        print(f"{self.name} says Woof!")
#Create objects.
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 4)
dog1.bark()
dog2.bark()
```
