Operator Overloading allows developers to define how operators such as `+`, `-`, `*`, `/`, and comparison operators behave when used with custom objects. This is achieved by overriding Python's **magic methods** (also called **dunder methods**).

```python

# Overloading the `+` Operator
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(
            self.x + other.x,
            self.y + other.y
        )

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1 + v2) 
```

```python
#Without `__repr__()`, printing an object displays its memory address.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Vector(
            self.x + other.x,
            self.y + other.y
        )
    def __sub__(self, other):
        return Vector(
            self.x - other.x,
            self.y - other.y
        )
    def __mul__(self, other):
        return Vector(
            self.x * other.x,
            self.y * other.y
        )
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1 + v2)
print(v1 - v2)
print(v1 * v2)
```
