#map function applies a given function to every item in an iterable (such as list, tuple, or set) and returns a map object

```python
#map function
#map(function, iterable)
def square(x):
    return x * x
print(square(10))
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = list(map(square, numbers))
print(result)
```

```python
#using map with lambda function
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = list(map(lambda x: x * x, numbers))
print(result)
```

```python
#map() with multiple iterables
num1 = [1, 2, 3]
num2 = [4, 5, 6]
added_numbers = list(map(lambda x, y: x + y, num1, num2))
print(added_numbers)
```

```python
#Converting Strings to Integers
str_numbers = ["1", "2", "3", "4", "5"]
int_numbers = list(map(int, str_numbers))
print(int_numbers)
```

```python
#Using Strings to uppercase
words = ["apple", "banana", "cherry"]
upper_words = list(map(str.upper, words))
print(upper_words)
```

```python
#Using map with dictionaries
people = [
    {"name": "Prav", "age": 27},
    {"name": "Kosa", "age": 28},
    {"name": "Rikku", "age": 30}
]
def get_name(person):
    return person["name"]
names = list(map(get_name, people))
print(names)
```
