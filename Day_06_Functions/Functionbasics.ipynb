```python
#Defining functions
def even_or_odd(num):
    """
    This function checks whether a number is even or odd.
    """
    if num % 2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")
even_or_odd(24) #Calling a function
```

```python
#Function parameters
def add(a, b):
    """
    Returns the sum of two numbers.
    """
    return a + b
result = add(5, 4)
print(result)
```

```python
#function with multiple parameters
def multiply(a, b, c):
    return a * b * c
print(multiply(2, 3, 4))
```

```python
#Default parameters
def greet(name="Prav"):
    print(f"Hello {name}")
greet()
greet("prav2")
```

```python
#Variable length arguments
#Sometimes we don't know how many arguments a function will receive.
#Positional Arguments (`*args`)
def print_numbers(*args):
    for number in args:
        print(number)
print_numbers(1,2,3,4,5,6,7,8,"prav")
```

```python
#Keyword Arguments (`**kwargs`)
#**kwargs allows a function to accept any number of keyword arguments.
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_details(
    name="Prav",
    age=25,
    country="India"
)
```

```python
#using positional arguments and keyword arguments together
def print_details(*args, **kwargs):
    for value in args:
        print(f"Positional Argument: {value}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print_details(
    1,
    2,
    3,
    name="Prav",
    age=27,
    country="India"
)
```

```python
#Return Statements
def multiply(a, b):
    return a * b
result = multiply(4, 5)
print(result)
#Returning Multiple Values
def multiply(a, b):
    return a * b, a
product, first_number = multiply(5, 6)
print(product)
print(first_number)
```
