```python
#lambda functions
#lambda arguments: expression
square = lambda x: x ** 2
print(square(5))
```

```python
#Regular vs lambda function
#regular
def addition(a, b):
    return a + b
#lambda
addition = lambda a, b: a + b
print(addition(6, 7))
print(type(addition))
```

```python
#lambda with multiple parameters
addition = lambda x, y, z: x + y + z
print(addition(1, 2, 3))
#lambda with boolean expression
even = lambda num: num % 2 == 0
print(even(12))
```

```python
#map function using lambda
#The map() function applies another function to every element in an iterable.
#map(function, iterable)
numbers = [1, 2, 3, 4, 5, 6]
def square(number):
    return number ** 2
result = list(map(square, numbers))
print(result)
#using lambda
numbers = [1, 2, 3, 4, 5, 6]
result = list(map(lambda num: num ** 2, numbers))
print(result)
```
