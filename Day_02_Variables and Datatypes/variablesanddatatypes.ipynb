Variables are used to store data

```python
#Variable initialization and declaration
a = 10;
print(type(a))
name = "string";
print(type(name))
```

Naming Conventions 
Variable name should be descriptive meaning it should clearly tell what data is stored inside
Variables should start with letter or underscore and can contain numbers, letters and underscores
variables are case sensitive

```python
#Naming convention
_isStudent = "True"
name2 = "prav"
first_name = "kosanam"
print(_isStudent)
print(name2)
print(first_name)
```

Type Checking 
We can check type of variable by using type()

```python
#Type Checking
a = 2.5
print(type(a))
isStudent="True"
print(type(isStudent))
last_name = "kosa"
print(type(last_name))
```

Type Conversion 
Type Conversion is converting from one type to another 
Ex: int to string
float to string

```python
#Type Conversion
age2 = 8
print(type(age2))
age2_str = str(age2)
print(type(age2_str))

# Converting String is not possible to int

student_name = "prav"
#student_name_int = int(student_name) 
#not possible
```

Dynamic typing
python allows the type of variable to change as program executes 

```python
a = 15;
print(type(a))
a="prav"
print(type(a))
a=15.5
print(type(a))
```

Taking input from the user by using input() 
Default user input type is string we have to convert whatever type we want

```python
# Taking input from the user
num1 = int(input("Enter the number 1"))
num2 = int(input("Enter the number 2"))
sum = num1+num2
print(sum)
```

Datatypes are classification of data which tell the compiler or interpreter how programmer intends
to use the data

They determine the types of operations that can be performed on the data, the values that the data
can take and the amount of memory needed to store the data

Importance of Datatypes in programming
Data types ensure that data is stored in an efficient way
They help in performing correct operations on data
proper use of datatypes can prevent errors and bugs in the program
