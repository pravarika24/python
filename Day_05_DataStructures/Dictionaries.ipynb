Dictionary
A dictionary is an unordered collection of items that stores data as key-value pairs
Each key is unique and immutable (such as strings, numbers, or tuples), while values can be of any data type.

```python
student = {
    "name": "prav",
    "age": 32,
    "grade": "A"
}
print(student)
print(type(student))
```

```python
#Creating dictionaries
empty_dict = {}
empty_dict = dict()
#Duplicate keys
student = {
    "name": "Prav",
    "age": 27,
    "name": "prav2"
}

print(student)
#Dictionary keys must be unique. If duplicate keys exist, the last value overwrites the previous one.
```

```python
#Accessing Dictionary Elements
student = {
    'name' : 'prav',
    'age':32,
    'grade' :'A'
}
print(student['grade']) #Using Square Brackets
print(student['name'])
print(student.get("grade"))#using get function
print(student.get("last_name"))#if the key is not present it will return none
print(student.get("last_name", "Not Available"))#Providing a default value if key is not present
```

```python
#Modifying Dictionaries
student["age"] = 33 #updating a value
student["address"] = "India"#adding a key
del student["grade"]#deleting a key
```

```python
#Dictionary methods
print(student.keys())#Get all keys
print(student.values())#Get all values
print(student.items())#Get all items
student.pop("age") #removes a key
student.clear()#clears the dictionary
```

```python
#Shallow copy 
#we have to use copy() method otherwise if any one thing in dictionary changes the other thing also changes
student = {
    "name": "Prav",
    "age": 32
}
student_copy = student
student["name"] = "Prav2"
print(student)
print(student_copy)
#Creating a shallow copy
student_copy = student.copy()
student["name"] = "Prav3"
print(student)
print(student_copy)
```

```python
#Nested Dictionaries
students = {
    "student1": {
        "name": "Rikku",
        "age": 32
    },
    "student2": {
        "name": "Prav",
        "age": 28
    }
}
print(students["student2"]["name"])
```

```python
for student_id, student_info in students.items():
    print(student_id)
    for key, value in student_info.items():
        print(f"{key}: {value}")
```

```python
#Dictionary Comprehensions
squares = {x: x**2 for x in range(5)}
#Conditional Dictionary Comprehension
evens = {x: x**2 for x in range(10) if x % 2 == 0}
print(evens)
```
