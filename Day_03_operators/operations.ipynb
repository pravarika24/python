```python
#Arithmetic operators
a = 15
b = 20
add_result = a+b #addition
sub_result = b-a #subtraction
multi_result = a*b #multiplication
div_result = a/b #division
floor_div_result = a//b #floor division(round off to floor)
modulus_result = a%b #modulus
exponent_result = a**b #exponentiation

print(add_result)
print(sub_result)
print(multi_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)
```

```python
#comparison operators
# == Equal to
a=10
b=20
a==b #comparing to identifier

str1 = "prav"
str2 = "kosanam"
str1 == str2
```

```python
# Not Equal to 

str1!=str2
```

```python
#greater than

num1 = 45
num2 = 55
print(num1>num2) #greater than
print(num1<num2) #less than
print(num1<=num2) #less than or equal to
print(num1>=num2) #greater than or equal to
```

```python
#logical operators - And, Not, OR
X = True
Y = False
result = X and Y
print(result)
result2 = X or Y 
print(result2)
result3 = not X #not is the reverse of the boolean value
print(result3)
```
