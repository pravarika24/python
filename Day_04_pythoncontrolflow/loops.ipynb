Loops
Loops allow us to execute the same block of code multiple times without rewriting it.

Python provides two types of loops:
- for loop
- while loop

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,6):
    print(i)
```

The third parameter is called the **step size**.

```python
for i in range(1,10,2):
    print(i)
```

If you use a negative step parameter, the first number should be higher and the second number should be lower

```python
for i in range (10,1,-2):
    print(i)
```

Iterating over string


```python
name = "pravarika"
for i in name:
    print(i)
```

while loop 
A `while` loop executes as long as the specified condition is **True**.

```python
num=0
while num<=5:
    print(num)
    num+=1
```

Loop Control Statements
break
continue
pass

Break
Break statement immediately terminates the loop

```python
for i in range(10):
    if i==5:
        break
    print("num is",i)
```

Continue
Continue statement skips the current iteration and moves to next iteration

```python
for i in range(10):
    if i %2==0:
        continue
    print(i)
```

pass
The pass statement does nothing
it acts as a placeholder

```python
for i in range(5):
    if i ==3:
        pass
    print(i)
```

Nested loop 
A nested loop is a loop inside another loop

```python
for i in range(4):
    for j in range(3):
        print(f" i = {i} and j = {j}")
```
