```python
# python is case sensitive

name = "prav"
Name = "Kosanam"

print(name)
print(Name)
```

python has indentation meaning other languages use {} to represent block of code but in python spacing(indentation) is important. It uses consistent indentation (PEP 8 recommends 4 spaces) for every block

```python
#This is to explain conditional statement to be precise if-else block
if True:
    print("this is true block")
else:
    print("this is false block")
print("this is out of block")
```

Line continuation -  You can use backslash (\) for line continuation

```python
total = 10 + 23 + 42+ 89 + \
      34 + 23

print(total)
```

Multiple statements in single line - we can write multiple statements in single line by dividing with ;

```python
x=7;y=12;z=x+y
print(z)
```
